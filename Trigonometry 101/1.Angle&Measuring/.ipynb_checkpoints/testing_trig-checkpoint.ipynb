```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
angle = np.arange(0,360)
sin = np.sin(angle)
cos = np.cos(angle)
# plt.plot(angle)
cir = plt.Circle((.5,.5),.3,color='b',fill=False)
fig,ax = plt.subplots()

plt.title('Circle 0.3 unit')
ax.set_aspect(1)
ax.add_artist(cir)
plt.show()
```

```python
theta = np.linspace(0,2*np.pi,100)
radius = 1
x = radius*np.cos(theta)
y = radius*np.sin(theta)

plt.figure(figsize=[8,8])
ax = plt.axes(aspect=1)
ax.plot(x,y)

init = [[0,-1],[0,1]]
ax.plot([[],[]],[[],[]])
plt.show()
```
